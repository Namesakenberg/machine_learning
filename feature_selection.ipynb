# drop duplicates

```python
# remove the duplicate columns:
duplicates = {} # dictionary to store duplicates
for i in X_train.columns:
  for j in X_train.columns:
    if i!=j and X_train[i].equals(X_train[j]):
      duplicates.setdefault(i, []).append(j)
```

```python
for key, values in duplicates.items():
    print(f"Column '{key}' is duplicated in: {values}")

# key : original column
# value : duplicate column
```

```python
# since every key (original column) has multiple values(duplicate columns) we would delete values and keep keys
for duplicate_cols in duplicates.values():
  for cols in duplicate_cols:
    X_train.drop(columns=cols , inplace = True,errors='ignore')
    X_test.drop(columns=cols , inplace = True,errors='ignore')
```

```python
# check the shape after dropping duplicates
print(X_train.shape)
print(X_test.shape)
```

# Variance Threshold

```python
from sklearn.feature_selection import VarianceThreshold
sel = VarianceThreshold(threshold=0.05)   # set a threshold
sel.fit(X_train)
```

```python
print(sum(sel.get_support()))

# sel.get support shows the true values in a particular column
```

```python
columns = X_train.columns[sel.get_support()]
```

```python
columns

# these are the 332 columns having variance above the threshold
```

```python
X_train = sel.transform(X_train)
X_test  = sel.transform(X_test)

X_train = pd.DataFrame(X_train,columns=columns)
X_test = pd .DataFrame(X_test , columns = columns)
```

```python
print(X_train.shape)
print(X_test.shape)
```

# drop columns on the basis of correlation coefficient

```python
corr_mat =X_train.corr()
```

```python
sns.heatmap(X_train.corr())
```

```python
columns = corr_mat.columns    # get the columns
columns_to_drop =[]           # array for the columns to be dropped
# for every column check its correlation with the other columns , if the correlation is > 0.95 drop the column

for i in range(len(columns)):
  for j in range(i+1 , len(columns)):
    if corr_mat.loc[columns[i],columns[j]] > 0.95:
      columns_to_drop.append(columns[j])

print(len(columns_to_drop))
```

```python
columns_to_drop = set(columns_to_drop)
print(len(columns_to_drop))
```

```python
X_train.drop(columns = columns_to_drop,axis=1,inplace = True,errors='ignore')
X_test.drop(columns = columns_to_drop , axis =1 ,inplace = True,errors='ignore')
```

```python
print(X_train.shape)
print(X_test.shape)
```

# Chi square test

since the chi square test can be only applied on the data where both the input and the output columns are categorical , using the titanic dataset

```python
titanic = pd.read_csv('/content/titanic.csv')
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
titanic = titanic[['Pclass','Sex','SibSp','Parch','Embarked','Survived']]
# select only the categorical columns as the input
```

```python
titanic.head()
```

```python
# find whether there is any relation between 'Sex' anf 'Survived' columns
```

```python
# since both the variables are categorical , make a contingency table(observed value)

ct = pd.crosstab(titanic['Survived'],titanic['Sex'],margins=True)
ct
```

```python
# make an expected table which is ALL(0) * all(female) / all(All) will give the expected value of the female(0)
# similarly calculate for all the values. and calculate sum of (observed-expected)^2/expected
# this value will be the chisquare statistic , now plot the chi square distribution for the n-1 dof (n = no. of input features)
# from the graph calculate the area beyond the chi sq statistic , it would be the p value
# if pvalue < 0.05 then reject the null hypothesis which was the column 'Sex' is related to 'survived'
```

```python
# making expected value cont tab
from scipy.stats import chi2_contingency
chi2_contingency(ct)
```

```python
# calculate for all the columns :
score = []

for feature in titanic.columns[:-1]:
  ct = pd.crosstab(titanic['Survived'],titanic[feature])  # crosstab of all the columns one by one against the survived column
  p_value = chi2_contingency(ct)[1] # first reading of the output is the p_value
  score.append(p_value)
```

```python
titanic.columns
```

```python
sns.barplot(score)
```

```python
# larger the  p value less important is the feature

# hence 3 i.e 'Parch' is the least important feature and 0 i.e 'Sex' is the most important feature
```
